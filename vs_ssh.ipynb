```python
# @title Colab for SSH
#setup colab_ssh
%pip install colab_ssh --upgrade
#vscode
%pip install -U ipykernel

from google.colab import drive, userdata
from colab_ssh import launch_ssh_cloudflared, init_git_cloudflared

#@markdown ###Global variables
repository_url='https://github.com/toddlack/colab.git' #@param{type:"string"  }
repository_branch='test' #@param{type:"string"}
repository_username='toddlack' #@param{type:"string"}
github_token_secret='github_token_todd' #@param{type:"string"}
cloudFlared_pw='testpw' #@param{type:"string"}

#connect to Google drive
drive.mount('/content/drive')

#launch ssh tunnel
launch_ssh_cloudflared(password=cloudFlared_pw,verbose=True,kill_other_processes=True)
# Optional: if you want to clone a Github or Gitlab repository
#repository_url="<PUT_YOUR_REPOSITORY_URL_HERE>"
github_token=userdata.get(github_token_secret)
init_git_cloudflared(repository_url,
                     personal_token=github_token,
                     branch=repository_branch,
                     username=repository_username
                     )
```
